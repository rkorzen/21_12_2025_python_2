## rekurencja

$$
n! = n \dot (n-1)!;0! = 1
$$


```python
def silnia(n):
    if n == 0:
        return 1
    return n * silnia(n-1)

silnia(5)
    
```

## dekoratory

```python
from functools import wraps

def bold(func):

    @wraps(func)
    def wrapper(*args, **kwargs):
        r = func(*args, **kwargs)
        return f"<b>{r}</b>"

    # wrapper = wraps(func)(wrapper)
    # wrapper.__doc__ = func.__doc__
    # wrapper.__name__ = func.__name__
    return wrapper

@bold
def text_a():
    """Zwraca tekst powitalny"""
    return "Hello world"


# text_a = bold(text_a)

@bold
def text_b(flag=False):
    if flag:
        return "Boom"
    return "Cos tam"

# text_b = bold(text_b)

@bold
def text_c(a, b, c):
    return a + b + c


# text_c = bold(text_c)

help(text_a)

assert text_a() == "<b>Hello world</b>"
assert text_b() == "<b>Cos tam</b>"
assert text_b(True) == "<b>Boom</b>"

assert text_c(1, 2, 3) == "<b>6</b>"
assert text_c("A", "B", "C") == "<b>ABC</b>"

```

```python
def foo(repeat=1):
    def decorator(func):
        def wrapper(*args, **kwargs):
            r = ""
            for i in range(repeat):
                r += func(*args, **kwargs)
            return r
        return wrapper

    return decorator

@foo(12)
def bar():
    print("10")
    return("10")

bar()
```
