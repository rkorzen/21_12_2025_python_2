## Zadanie 2: Analiza danych sprzedaowych

Dane sprzedaowe zapisane s jako lista sownikw. Kady element zawiera:

* nazw produktu,
* kategori produktu,
* liczb sprzedanych sztuk.

```python
sales = [
    {"product": "Laptop", "category": "Electronics", "quantity": 15},
    {"product": "Mouse", "category": "Electronics", "quantity": 120},
    {"product": "Keyboard", "category": "Electronics", "quantity": 85},
    {"product": "Monitor", "category": "Electronics", "quantity": 0},
    {"product": "Desk", "category": "Furniture", "quantity": 20},
    {"product": "Chair", "category": "Furniture", "quantity": 55},
    {"product": "Lamp", "category": "Furniture", "quantity": 35},
    {"product": "Notebook", "category": "Office", "quantity": 200},
    {"product": "Pen", "category": "Office", "quantity": 350},
    {"product": "Stapler", "category": "Office", "quantity": 40},
]
```

### Twoim zadaniem jest:

1. Obliczy czn sprzeda (sum `quantity`) dla kadej kategorii.
2. Wypisa kategori, ktra ma **najwysz czn sprzeda**.
3. Wypisa wszystkie produkty, ktrych sprzeda jest **wiksza ni rednia sprzeda w ich kategorii**.
4. Sprawdzi, czy istniej produkty, ktrych sprzeda wynosi `0`, i wypisa ich nazwy (lub informacj, e takich produktw nie ma).


````python
sales = [
    {"product": "Laptop", "category": "Electronics", "quantity": 15},
    {"product": "Mouse", "category": "Electronics", "quantity": 120},
    {"product": "Keyboard", "category": "Electronics", "quantity": 85},
    {"product": "Monitor", "category": "Electronics", "quantity": 0},
    {"product": "Desk", "category": "Furniture", "quantity": 20},
    {"product": "Chair", "category": "Furniture", "quantity": 55},
    {"product": "Lamp", "category": "Furniture", "quantity": 35},
    {"product": "Notebook", "category": "Office", "quantity": 200},
    {"product": "Pen", "category": "Office", "quantity": 350},
    {"product": "Stapler", "category": "Office", "quantity": 40},
]

# --------------------------------------------------
# 1. czna sprzeda dla kadej kategorii
# --------------------------------------------------

category_totals = {}

for item in sales:
    category = item["category"]
    quantity = item["quantity"]

    if category not in category_totals:
        category_totals[category] = 0

    category_totals[category] += quantity

print("czna sprzeda wg kategorii:")
for category, total in category_totals.items():
    print(f"- {category}: {total}")

# --------------------------------------------------
# 2. Kategoria z najwysz sprzeda
# --------------------------------------------------
````

````python
top_category = None
top_value = 0

for category, total in category_totals.items():
    if total > top_value:
        top_value = total
        top_category = category

print(f"\nKategoria z najwysz sprzeda: {top_category}")
````

````python
max(category_totals.items(), key=lambda x: x[1])
````

````python
# --------------------------------------------------
# 3. Produkty powyej redniej w swojej kategorii
# --------------------------------------------------

category_counts = {}

for item in sales:
    category = item["category"]

    if category not in category_counts:
        category_counts[category] = 0

    category_counts[category] += 1

category_averages = {}

for category in category_totals:
    category_averages[category] = (
        category_totals[category] / category_counts[category]
    )

print("\nProdukty powyej redniej w swojej kategorii:")
for item in sales:
    category = item["category"]
    quantity = item["quantity"]

    if quantity > category_averages[category]:
        print(f"- {item['product']} ({category}, {quantity})")

# --------------------------------------------------
# 4. Produkty ze sprzeda rwn 0
# --------------------------------------------------

zero_sales = []

for item in sales:
    if item["quantity"] == 0:
        zero_sales.append(item["product"])

if zero_sales:
    print("\nProdukty ze sprzeda 0:")
    for product in zero_sales:
        print(f"- {product}")
else:
    print("\nBrak produktw ze sprzeda 0")
````

## Zadanie 3: Analiza aktywnoci uytkownikw (login / logout + czas)

Dane aktywnoci uytkownikw zapisane s jako lista zdarze.
Kade zdarzenie zawiera:

* identyfikator uytkownika (`user_id`),
* typ zdarzenia (`"login"`, `"logout"`, `"purchase"`),
* czas zdarzenia (`time`)  liczba cakowita, rosnca w skali caego systemu,
* kwot (`amount`)  dla zdarze innych ni `"purchase"` zawsze `0`.

### Dane wejciowe

```python
events = [
    {"user_id": "u1", "type": "login",    "time": 1,  "amount": 0},
    {"user_id": "u1", "type": "purchase", "time": 5,  "amount": 120},
    {"user_id": "u1", "type": "logout",   "time": 10, "amount": 0},

    {"user_id": "u2", "type": "login",    "time": 12, "amount": 0},
    {"user_id": "u2", "type": "logout",   "time": 18, "amount": 0},

    {"user_id": "u3", "type": "login",    "time": 20, "amount": 0},
    {"user_id": "u3", "type": "purchase", "time": 25, "amount": 50},
    {"user_id": "u3", "type": "purchase", "time": 30, "amount": 75},
    {"user_id": "u3", "type": "logout",   "time": 40, "amount": 0},

    {"user_id": "u4", "type": "login",    "time": 45, "amount": 0},
    {"user_id": "u4", "type": "logout",   "time": 50, "amount": 0},
]
```

---

### Twoim zadaniem jest:

1. Zliczy, ilu **unikalnych uytkownikw** wystpuje w danych.
2. Obliczy **czn warto zakupw** dla kadego uytkownika.
3. Obliczy **czny czas spdzony w systemie** przez kadego uytkownika
   (rnica `logout.time - login.time`).
4. Wypisa uytkownikw, ktrzy:

   * wykonali co najmniej jeden `"login"`,
   * ale **nie wykonali adnego `"purchase"`.
5. Wypisa uytkownika, ktry spdzi **najwicej czasu w systemie**.

````python
events = [
    {"user_id": "u1", "type": "login",    "time": 1,  "amount": 0},
    {"user_id": "u1", "type": "purchase", "time": 5,  "amount": 120},
    {"user_id": "u1", "type": "logout",   "time": 10, "amount": 0},

    {"user_id": "u2", "type": "login",    "time": 12, "amount": 0},
    {"user_id": "u2", "type": "logout",   "time": 18, "amount": 0},

    {"user_id": "u3", "type": "login",    "time": 20, "amount": 0},
    {"user_id": "u3", "type": "purchase", "time": 25, "amount": 50},
    {"user_id": "u3", "type": "purchase", "time": 30, "amount": 75},
    {"user_id": "u3", "type": "logout",   "time": 40, "amount": 0},

    {"user_id": "u4", "type": "login",    "time": 45, "amount": 0},
    {"user_id": "u4", "type": "logout",   "time": 50, "amount": 0},
]


# --------------------------------------------------
# 1. Unikalni uytkownicy
# --------------------------------------------------

users = set()

for event in events:
    users.add(event["user_id"])

print(f"Liczba unikalnych uytkownikw: {len(users)}")

# --------------------------------------------------
# 2. czna warto zakupw
# --------------------------------------------------

purchase_totals = {}

for event in events:
    user = event["user_id"]

    if user not in purchase_totals:
        purchase_totals[user] = 0

    if event["type"] == "purchase":
        purchase_totals[user] += event["amount"]

print("\nczna warto zakupw:")
for user, total in purchase_totals.items():
    print(f"- {user}: {total}")

# --------------------------------------------------
# 3. Czas spdzony w systemie
# --------------------------------------------------

login_times = {}
time_spent = {}

for event in events:
    user = event["user_id"]

    if user not in time_spent:
        time_spent[user] = 0

    if event["type"] == "login":
        login_times[user] = event["time"]

    elif event["type"] == "logout":
        session_time = event["time"] - login_times[user]
        time_spent[user] += session_time

print("\nCzas spdzony w systemie:")
for user, total_time in time_spent.items():
    print(f"- {user}: {total_time}")

# --------------------------------------------------
# 4. Login bez zakupu
# --------------------------------------------------

logins = {}
purchases = {}

for event in events:
    user = event["user_id"]

    if user not in logins:
        logins[user] = 0
        purchases[user] = 0

    if event["type"] == "login":
        logins[user] += 1
    elif event["type"] == "purchase":
        purchases[user] += 1

print("\nUytkownicy z loginem, ale bez zakupu:")
for user in logins:
    if logins[user] > 0 and purchases[user] == 0:
        print(f"- {user}")

# --------------------------------------------------
# 5. Najduszy czas w systemie
# --------------------------------------------------

top_user = None
max_time = 0

for user, total_time in time_spent.items():
    if total_time > max_time:
        max_time = total_time
        top_user = user

print(f"\nNajwicej czasu w systemie spdzi: {top_user} ({max_time})")
````
