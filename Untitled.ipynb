### Wymagania techniczne

* zastosuj **funkcje jako obiekty pierwszej klasy**:

  * funkcje modyfikujÄ…ce prowizjÄ™ przekazywane jako argumenty,
* uÅ¼yj:

````python

from collections import defaultdict
from typing import Callable, Iterable, Dict, List


# =========================
# DANE
# =========================

sales = [
    {"seller": "Anna", "region": "EU", "client_type": "B2B", "amount": 120_000},
    {"seller": "Anna", "region": "EU", "client_type": "B2C", "amount": 15_000},
    {"seller": "Anna", "region": "US", "client_type": "B2B", "amount": 90_000},

    {"seller": "Bartek", "region": "US", "client_type": "B2C", "amount": 8_000},
    {"seller": "Bartek", "region": "EU", "client_type": "B2C", "amount": 22_000},

    {"seller": "Celina", "region": "EU", "client_type": "B2B", "amount": 250_000},
    {"seller": "Celina", "region": "ASIA", "client_type": "B2B", "amount": 60_000},

    {"seller": "Daniel", "region": "ASIA", "client_type": "B2C", "amount": 12_000},
    {"seller": "Daniel", "region": "EU", "client_type": "B2C", "amount": 9_000},
]


# =========================
# REGUÅY BIZNESOWE
# =========================

def base_rate(amount: float) -> float:
    return 0.05


def region_modifier(amount: float, **context) -> float:
    rates = {
        "EU": 0.01,
        "US": 0.02,
        "ASIA": 0.03,
    }
    return rates.get(context["region"], 0.0)


def client_type_modifier(amount: float, **context) -> float:
    return 0.02 if context["client_type"] == "B2B" else 0.0


def high_value_bonus(
    amount: float,
    *,
    threshold: float = 100_000,
    bonus: float = 0.01,
    **context
) -> float:
    return bonus if amount > threshold else 0.0


# =========================
# SILNIK (FCO)
# =========================

BaseFn = Callable[[float], float]
ModifierFn = Callable[..., float]


def calculate_commission(
    *,
    amount: float,
    base_fn: BaseFn,
    modifiers: Iterable[ModifierFn],
    **context
) -> float:
    rate = base_fn(amount)

    for modifier in modifiers:
        rate += modifier(amount, **context)

    return amount * rate


def calculate_commissions(
    sales: List[Dict],
    *,
    base_fn: BaseFn,
    modifiers: Iterable[ModifierFn]
) -> Dict[str, float]:
    result = defaultdict(float)

    for sale in sales:
        commission = calculate_commission(
            amount=sale["amount"],
            base_fn=base_fn,
            modifiers=modifiers,
            region=sale["region"],
            client_type=sale["client_type"],
        )
        result[sale["seller"]] += commission

    return dict(result)


# =========================
# STRATEGIA
# =========================

commission_strategy = [
    region_modifier,
    client_type_modifier,
    high_value_bonus,
]


# =========================
# URUCHOMIENIE
# =========================

if __name__ == "__main__":
    commissions = calculate_commissions(
        sales,
        base_fn=base_rate,
        modifiers=commission_strategy,
    )

    for seller, value in commissions.items():
        print(f"{seller}: {value:.2f} PLN")
````
